# Hyperparameter Optimization Methods Review

In this notebook we will assess how different hyperparamet optimization methods perform on a specific dataset. Methods analyzed cover naive approaches such as performing random search or grid search on the hyperparameter space but also include a meta-heuristic method in the genetic algorithms and a statistical approach in the Bayesian Optimization.

```python
import pandas as pd
import numpy as np
import os

# Plotting imports
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution related imports
from scipy.stats import randint
import sklearn_genetic

# Model imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Metric imports
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

# Search imports
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn_genetic import GASearchCV
from skopt import BayesSearchCV
```

```python
# Reading the data file
columns = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points']

for i in range(4):
    columns.append('Wilderness_Area' + str(i))

for i in range(40):
    columns.append('soil_type' + str(i))

columns.append('cover_type')

os.chdir(r'C:\Users\nicol\Documents\Mestrado\Mat√©rias\Expert Systems\Semester project')
df = pd.read_csv("covtype.data", header=None, names = columns)

df['cover_type'].value_counts()
```

```python
# Create the train test split 

y = df['cover_type']
df.drop(columns = ['cover_type'], inplace = True)

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25)
```

# Random Search

```python
# Random Search

# Define hyperparameter space to be explored
params_RS = {'n_estimators':randint(10, 100), 'criterion':['gini', 'entropy'], 'min_samples_split': randint(2, 20)}

# Define model and fit
rfc = RandomForestClassifier()
clf = RandomizedSearchCV(rfc, params_RS, n_iter = 50, verbose = 2)
clf.fit(X_train, y_train)

history = clf.cv_results_
pred = clf.predict(X_test)

# Report best set of hyperparameters
print(clf.best_estimator_, clf.best_score_, clf.best_params_)
```

```python
# Plot Results
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Scores based on hyper parameters')

df = pd.DataFrame.from_dict(history['params'])
df['scores'] = history['mean_test_score']

df.to_csv('Results Random search.csv', sep = ';')

sns.scatterplot(x = 'min_samples_split', y = 'n_estimators', size = 'scores', sizes = (5,600), hue = 'criterion', data = df, ax = ax)
ax.legend(bbox_to_anchor=(1.1, 1.02))
```

# Grid Search

```python
# Grid Search

params_GS = {'n_estimators':[10, 30, 50, 75, 100], 'criterion':['gini', 'entropy'], 'min_samples_split': [2, 5, 10, 20]}
rfc = RandomForestClassifier()
clf = GridSearchCV(rfc, params_GS, verbose = 2)
clf.fit(X_train, y_train)

history = clf.cv_results_
pred = clf.predict(X_test)

print(clf.best_estimator_, clf.best_score_, clf.best_params_)
```

```python
#Plot results
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Scores based on hyper parameters')

df = pd.DataFrame.from_dict(history['params'])
df['scores'] = history['mean_test_score']

df.to_csv('Results grid search.csv', sep = ';')

sns.scatterplot(x = 'min_samples_split', y = 'n_estimators', size = 'scores', sizes = (5,600), hue = 'criterion', alpha = 0.5, data = df, ax = ax)
ax.legend(bbox_to_anchor=(1.1, 1.02))   
```

# Genetic Algorithm

```python
# Genetic Algorithm

# Define hyperparameter space
n_est = sklearn_genetic.space.Integer(10, 100)
min_samples_split = sklearn_genetic.space.Integer(2, 20)
criterion = sklearn_genetic.space.Categorical(['gini', 'entropy'])

# Define model and fit
params_GA = {'n_estimators':n_est, 'criterion':criterion, 'min_samples_split': min_samples_split}
rfc = RandomForestClassifier()
evolved_estimator = GASearchCV(estimator=rfc,
                               scoring='accuracy',
                               population_size=10,
                               generations=10,
                               tournament_size=5,
                               elitism=True,
                               crossover_probability=0.8,
                               mutation_probability=0.1,
                               param_grid=params_GA,
                               criteria='max',
                               algorithm='eaMuPlusLambda',
                               n_jobs=-1,
                               verbose=True,
                               keep_top_k=4)

evolved_estimator.fit(X_train, y_train)

history = evolved_estimator.cv_results_
evolution = evolved_estimator.history

pred = evolved_estimator.predict(X_test)

multilabel_confusion_matrix(y_test, pred)
accuracy_score(y_test, pred)
f1_score(y_test, pred, average=None)

print(evolved_estimator.best_estimator_, evolved_estimator.best_score_, evolved_estimator.best_params_)
```

```python
# Plot results
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Scores based on hyper parameters')

df = pd.DataFrame.from_dict(history['params'])
df['scores'] = history['mean_test_score']# - np.min(history['mean_test_score']))/(np.max(history['mean_test_score'] - np.min(history['mean_test_score'])))

df.to_csv('Results GA 1.csv', sep = ';')

sns.scatterplot(x = 'min_samples_split', y = 'n_estimators', size = 'scores', sizes = (5,600), hue = 'criterion', data = df, ax = ax)
ax.legend(bbox_to_anchor=(1.1, 1.02))
```

```python
# Plot evolution of the fitnesses obtained
df_evo = pd.DataFrame.from_dict(evolution)

df_evo.to_csv('Results GA 2.csv', sep = ';')

fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Evolution across generations')

sns.scatterplot(x = np.arange(len(df_evo)), y = df_evo['fitness_max'], ax = ax)
```

# Bayesian Optimization

```python
# Bayesian Optimization approach

params_BS = {'n_estimators':np.arange(10,100,1), 'criterion':['gini', 'entropy'], 'min_samples_split':np.arange(2, 20, 1)}
rfc = RandomForestClassifier()
clf = BayesSearchCV(rfc, search_spaces=params_BS, verbose = 2)

clf.fit(X_train, y_train)
history = clf.cv_results_

pred = clf.predict(X_test)

multilabel_confusion_matrix(y_test, pred)
accuracy_score(y_test, pred)
f1_score(y_test, pred, average=None)

print(clf.best_estimator_, clf.best_score_, clf.best_params_)
```

```python
# Plot results
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Scores based on hyper parameters')

df = pd.DataFrame.from_dict(history['params'])
df['scores'] = history['mean_test_score']# - np.min(history['mean_test_score']))/(np.max(history['mean_test_score'] - np.min(history['mean_test_score'])))

df.to_csv('Results Bayesian Search.csv', sep = ';')

sns.scatterplot(x = 'min_samples_split', y = 'n_estimators', size = 'scores', sizes = (5,600), hue = 'criterion', data = df, ax = ax)
ax.legend(bbox_to_anchor=(1.1, 1.02))
```

```python
# Plot Evolution
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Evolution across generations')


sns.scatterplot(x = np.arange(len(df)), y = df['scores'], ax = ax)
```
