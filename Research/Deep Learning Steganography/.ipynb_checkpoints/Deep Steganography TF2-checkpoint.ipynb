# Deep Learning Steganography

In this notebook we re-implement the algorithm proposed by http://www.esprockets.com/papers/nips2017.pdf updating the code to the syntax of TensorFlow 2. 

Because we have multiple networks that should be tracked and trained independently, we use the a functional approach to building TensorFlow models. 

Finally, this code can be run on GPU's for better performance, but the dependencies on the environment.yml file found in this folder should be used.

```python
import numpy as np
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

import pydot
import pydotplus
import graphviz
from tensorflow.keras.utils import plot_model

import glob
import time
import os
from PIL import Image,ImageOps
import random

import time
from datetime import datetime
from os.path import join
```

```python
keras.backend.clear_session()
```

```python
# Setting User Parameters

TRAIN_PATH_1 = 'data/train/cifar-10 - full'
TEST_PATH = '/data/train/cifar-10 - test'
LOGS_Path = "../logs_new/"
CHECKPOINTS_PATH = './checkpoints/'
SCALE_TO = 224
TRAINING_SIZE = 4*100
EPOCHS = 10

BATCH_SIZE = 2
LEARNING_RATE = .0001
BETA = .75

EXP_NAME = f"beta_{BETA}"

files_list = glob.glob(join(TRAIN_PATH_1,"**/*"))
files_list_test = glob.glob(join(TEST_PATH,"**/*"))
```

```python
# Normalize batch of images
def normalize_batch(imgs):
    return (imgs -  np.array([0.485, 0.456, 0.406])) /np.array([0.229, 0.224, 0.225]) 

# Denormalize batch of images
def denormalize_batch(imgs,should_clip=True):
    imgs= (imgs * np.array([0.229, 0.224, 0.225])) + np.array([0.485, 0.456, 0.406])
    if should_clip:
        imgs= np.clip(imgs,0,1)
    return imgs

# Get a specific batch of images to use in prediction
def get_img_batch(files_list,batch_size=32,size=(SCALE_TO,SCALE_TO),should_normalise=True):
   
    batch_cover = []
    batch_secret = []

    for i in range(batch_size):
        #Choose random image
        img_secret_path = random.choice(files_list)
        img_cover_path = random.choice(files_list)
        
        #Convert random image to RGB
        img_secret = Image.open(img_secret_path).convert("RGB")
        img_cover = Image.open(img_cover_path).convert("RGB")
        
        #Crop image to the size the NN is ready to deal with
        img_secret = np.array(ImageOps.fit(img_secret,size),dtype=np.float32)
        img_cover = np.array(ImageOps.fit(img_cover,size),dtype=np.float32)
        
        # Make pixel values between 0 and 1
        img_secret /= 255.
        img_cover /= 255.
        
        # Append images to the list of images that are going to be in the batch
        batch_cover.append(img_cover)
        batch_secret.append(img_secret)

    batch_cover,batch_secret = np.array(batch_cover), np.array(batch_secret)
    
    if should_normalise:
        batch_cover = normalize_batch(batch_cover)
        batch_secret = normalize_batch(batch_secret)

    return batch_cover, batch_secret
```

```python
# Get the a "full dataset" of images, used in training
def get_images(files_list, number_of_images, offset, size=(SCALE_TO,SCALE_TO), should_normalise=True):
   
    batch_cover = []
    batch_secret = []

    for i in range(offset, int(number_of_images/2) + offset):
        #Choose random image
        img_secret_path = files_list[i] # random.choice(files_list)
        img_cover_path = files_list[i + int(number_of_images/2)] # random.choice(files_list)
        
        #Convert random image to RGB
        img_secret = Image.open(img_secret_path).convert("RGB")
        img_cover = Image.open(img_cover_path).convert("RGB")
        
        #Crop image to the size the NN is ready to deal with
        img_secret = np.array(ImageOps.fit(img_secret,size),dtype=np.float64)
        img_cover = np.array(ImageOps.fit(img_cover,size),dtype=np.float64)
        
        # Make pixel values between 0 and 1
        img_secret /= 255.
        img_cover /= 255.
        
        # Append images to the list of images that are going to be in the batch
        batch_cover.append(img_cover)
        batch_secret.append(img_secret)
    
    
    batch_cover,batch_secret = np.array(batch_cover), np.array(batch_secret)
    
    if should_normalise:
        batch_cover = normalize_batch(batch_cover)
        batch_secret = normalize_batch(batch_secret)

    return batch_cover, batch_secret
```

```python
# Define the preparation network graph
def get_prep_net(secret_tensor):
    
    scope_name_1 = 'prep_net'
    
    scope_name_2 = 'conv_3x3'
    # Convolutional layers with kernel size 3
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "1",activation=tf.nn.relu) (secret_tensor)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "2",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "3",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "4",activation=tf.nn.relu) (conv_3x3)
    
    scope_name_2 = 'conv_4x4'
    # Convolutional layers with kernel size 4
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "1",activation=tf.nn.relu) (secret_tensor)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "2",activation=tf.nn.relu) (conv_4x4)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "3",activation=tf.nn.relu) (conv_4x4)  
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "4",activation=tf.nn.relu) (conv_4x4)  
    
    scope_name_2 = 'conv_5x5'
    # Convolutional layers with kernel size 5
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "1",activation=tf.nn.relu) (secret_tensor)
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "2",activation=tf.nn.relu) (conv_5x5)        
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "3",activation=tf.nn.relu) (conv_5x5)   
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' + "4",activation=tf.nn.relu) (conv_5x5)   
    
    # Concatenation of the convolutions
    concat_1 = tf.concat([conv_3x3,conv_4x4, conv_5x5],axis=3,name=scope_name_1 + '/' +'concat_1')
    
    # Another batch of convolutions
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' +"final_5x5",activation=tf.nn.relu) (concat_1)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' +"final_4x4",activation=tf.nn.relu) (concat_1)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' +"final_3x3",activation=tf.nn.relu) (concat_1)
    
    # Final concatenation and casting of the prep network output tensor
    concat_final_1 = tf.concat([conv_5x5,conv_4x4,conv_3x3],axis=3,name=scope_name_1 + '/' +'concat_final_1')
    concat_final_1 = tf.cast(concat_final_1, dtype = 'float64', name = scope_name_1 + '/' +'cast_final_1')
    
    # Define model graph
    prep_net = keras.Model(
    inputs=[secret_tensor],
    outputs=[concat_final_1],
    )
    
    return prep_net

''' Hiding network, similar structure to the prep net, but with concatenation of the prep output and the 
    Cover tensor. Multiple convolutional layers, concatenation of the convoluted results and then
    a final set of convolutions before a final concatenation '''
def get_hide_net(cover_tensor, prep_output):
    
    scope_name_1 = 'hiding_net'
    concat_input = tf.concat([cover_tensor, prep_output],axis=3,name=scope_name_1 + '/' +'images_features_concat')
    
    scope_name_2 = '3x3_conv_branch'
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (concat_input)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_3x3)
    
    scope_name_2 = '4x4_conv_branch'
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (concat_input)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_4x4)  
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_4x4) 
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_4x4) 

    scope_name_2 = '5x5_conv_branch'
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (concat_input)
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_5x5)    
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_5x5)
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_5x5)

    concat_2 = tf.concat([conv_3x3,conv_4x4,conv_5x5],axis=3,name=scope_name_1 + '/' +'concat_1')

    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' +"final_5x5",activation=tf.nn.relu) (concat_2)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' +"final_4x4",activation=tf.nn.relu) (concat_2)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' +"final_3x3",activation=tf.nn.relu) (concat_2)

    concat_final_2 = tf.concat([conv_5x5,conv_4x4,conv_3x3],axis=3,name=scope_name_1 + '/' +'concat_final')
    output_hide = layers.Conv2D(filters=3,kernel_size=1,padding='same',name=scope_name_1 + '/' +'output') (concat_final_2)
    
    hide_net = keras.Model(
    inputs=[cover_tensor, prep_output],
    outputs=[output_hide],
    )
    
    return hide_net

def get_noise_layer(output_hide, std = 0.1):
    output_noise = output_hide + tf.random.normal(shape=tf.shape(output_hide), mean=0.0, stddev=std, dtype=tf.float64)
    
    noise_layer = keras.Model(
    inputs=[output_hide],
    outputs=[output_noise],
    )
    
    return noise_layer
''' Reveal network, similar structure to the prep net. Multiple convolutional layers, concatenation of
    the convoluted results and then a final set of convolutions before a final concatenation '''
def get_reveal_net(output_noise):
    
    scope_name_1 = 'reveal_net'
    scope_name_2 = '3x3_conv_branch'
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (output_noise)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_3x3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_3x3)
    

    scope_name_2 = '4x4_conv_branch'
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (output_noise)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_4x4)       
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_4x4)     
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_4x4)     


    scope_name_2 = '5x5_conv_branch'
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"1",activation=tf.nn.relu) (output_noise)
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"2",activation=tf.nn.relu) (conv_5x5)    
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"3",activation=tf.nn.relu) (conv_5x5)  
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' + scope_name_2 + '/' +"4",activation=tf.nn.relu) (conv_5x5)  
    
    concat_3 = tf.concat([conv_3x3,conv_4x4,conv_5x5],axis=3,name=scope_name_1 + '/' +'concat_1')
    
    conv_5x5 = layers.Conv2D(filters=50,kernel_size=5,padding='same',name=scope_name_1 + '/' +"final_5x5",activation=tf.nn.relu) (concat_3)
    conv_4x4 = layers.Conv2D(filters=50,kernel_size=4,padding='same',name=scope_name_1 + '/' +"final_4x4",activation=tf.nn.relu) (concat_3)
    conv_3x3 = layers.Conv2D(filters=50,kernel_size=3,padding='same',name=scope_name_1 + '/' +"final_3x3",activation=tf.nn.relu) (concat_3)
    
    concat_final_3 = tf.concat([conv_5x5,conv_4x4,conv_3x3],axis=3,name=scope_name_1 + '/' +'concat_final')
    
    output_reveal = layers.Conv2D(filters=3,kernel_size=1,padding='same',name=scope_name_1 + '/' +'output') (concat_final_3)
    
    reveal_net = keras.Model(
    inputs=[output_noise],
    outputs=[output_reveal],
    )
    
    return reveal_net

```

```python
# Get input tensors for the models. This is required by TensorFlow 
secret_input = tf.keras.layers.Input(shape = (SCALE_TO, SCALE_TO, 3), dtype='float64', batch_size = BATCH_SIZE)
cover_input = tf.keras.layers.Input(shape = (SCALE_TO, SCALE_TO, 3), dtype='float64', batch_size = BATCH_SIZE)

# Get prep net and generate an output tensor
prep_net = get_prep_net(secret_input)
output_prep = prep_net(secret_input)

# Use prep net output as input. Also add the secret image as an input and define the hide network
output_prep_as_input = tf.keras.layers.Input(shape = output_prep.shape[1:], dtype='float64', batch_size = BATCH_SIZE)
hide_net = get_hide_net(cover_input, output_prep_as_input)
output_hide = hide_net([cover_input, output_prep])

# Get noise layer and create graph
output_hide_as_input = tf.keras.layers.Input(shape = output_hide.shape[1:], dtype='float64', batch_size = BATCH_SIZE)
noise_layer = get_noise_layer(output_hide_as_input)
output_noise = noise_layer(output_hide)
output_noise

# Get reveal network and create graph
output_noise_as_input = tf.keras.layers.Input(shape = output_noise.shape[1:], dtype='float64', batch_size = BATCH_SIZE)
reveal_net = get_reveal_net(output_noise_as_input)
output_reveal = reveal_net(output_noise)

# Define the full model as the done in the paper and compile
hide_model = keras.Model(inputs=[secret_input, cover_input], outputs=[output_hide])
full_model =  keras.Model(inputs=[secret_input, cover_input], outputs=[output_reveal, output_hide])

full_model.compile(optimizer='adam', loss = [tf.keras.losses.MeanSquaredError(), tf.keras.losses.MeanSquaredError()],  loss_weights = [1,BETA])
```

```python
# Get image dataset
covers,secrets = get_images(files_list=files_list, number_of_images = TRAINING_SIZE, offset = 0)

# Train model
full_model.fit(x = (secrets, covers), y = (secrets, covers), batch_size =  BATCH_SIZE, epochs = EPOCHS, verbose = 1, validation_split = 0.1)
```

```python
# Get test images and predict the results  
covers,secrets = get_img_batch(files_list=files_list, batch_size = 1)
results_1 = full_model.predict(x = (secrets, covers), batch_size = BATCH_SIZE)
```
